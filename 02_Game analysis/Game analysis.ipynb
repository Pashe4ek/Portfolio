Для того, чтобы заполнить часть пропусков найдем самые популярные рейтинги на всех платформах.

````python
data.loc[mask, 'rating'] =  data.loc[mask, 'genre'].map(popular_genre_rating)
````

### Проверка гипотезы

<b>Средние пользовательские рейтинги платформ Xbox One и PC одинаковые; 

Перед проверкой, заполним пропуски. Так как нам нужно сравнить средние значения заполним пропуски средним значением. 

```
H_0: Средние пользовательские рейтинги платформ Xbox One ('Xone') и PC ('PC') равны =.
H_a: Средние пользовательские рейтинги платформ Xbox One ('Xone') и PC ('PC') не равны ≠.
alpha = 0.05
```

````python
results = st.ttest_ind(
    df[df['platform'] == 'XOne']['user_score'].dropna(),
    df[df['platform'] == 'PC']['user_score'].dropna(),
    equal_var=False
)
# results = вызов метода для проверки гипотезы
alpha = .05
# alpha = задайте значение уровня значимости
print(results.pvalue)
# вывод значения p-value на экран 
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
# условный оператор с выводом строки с ответом
````

Не удалось отвергнуть гипотезу, что средние пользовательские рейтинги платформ Xbox One ('Xone') и PC ('PC')  значит средние пользовательские рейтинги этих платформ статистически значимо равны

<b> Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

```
H_0: Средние пользовательские рейтинги жанров Action и Sports равны =
H_a: Средние пользовательские рейтинги жанров Action и Sports не равны ≠.
alpha = 0.05
```

````python
results = st.ttest_ind(
    df[df['genre'] == 'Action']['user_score'].dropna(),
    df[df['genre'] == 'Sports']['user_score'].dropna(),
    equal_var=False
)
# results = вызов метода для проверки гипотезы
alpha = .05
# alpha = задайте значение уровня значимости
print(results.pvalue)
# вывод значения p-value на экран 
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
# условный оператор с выводом строки с ответом
````

Удалось отвергнуть гипотезу, что средние пользовательские рейтинги жанров Action и Sports равны =, значит средние пользовательские рейтинги этих жанров статистически значимо различаются
